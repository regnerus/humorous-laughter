```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb

from matplotlib import cm
# sb.set(style="white", palette="muted")

def set_style():
    # This sets reasonable defaults for font size for
    # a figure that will go in a paper
    sb.set_context("paper")
    
    # Set the font to be serif, rather than sans
    sb.set(font='serif')
    
    # Make the background white, and specify the
    # specific font family
    sb.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
    })
    
    sb.set_palette("bright")

set_style()

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as ms
from matplotlib.colors import ListedColormap
ms.use('seaborn-bright')
%matplotlib inline

import librosa
import librosa.display
import IPython.display

import textgrid as tg
```

```python
file = "1_3c_1A_B"

wav_pathname = "data/"+file+".wav"

y, sr = librosa.load(wav_pathname)

textgrid = tg.TextGrid.fromFile("data/"+file+".TextGrid")
humorous_tier = textgrid[1]

# Let's make and display a mel-scaled power (energy-squared) spectrogram
S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)

# Convert to log scale (dB). We'll use the peak power as reference.
log_S = librosa.amplitude_to_db(S, ref=np.max)

# Make a new figure
plt.figure(figsize=(12,4))

# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=sr, cmap=cm.jet, x_axis='time', y_axis='mel')
marks = ['H', 'N']
for interval in humorous_tier: #humorous tier
    if interval.mark in marks:
        start = int(interval.minTime)
        end = int(interval.maxTime)
#         print(start,end)
        if(interval.mark == marks[0]):
            plt.axvspan(start, end, alpha=0.25, facecolor ='white', label='Humorous')
        else:
            plt.axvspan(start, end, alpha=0.25, facecolor ='black', label='Non-Humorous')

# Put a descriptive title on the plot
plt.title('mel power spectrogram')

# draw a color bar
plt.colorbar(format='%+02.0f dB')

# Make the figure layout compact
plt.tight_layout()
```

```python
# Next, we'll extract the first 13 Mel-frequency cepstral coefficients (MFCCs)
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=13)

# Padding first and second deltas
delta_mfcc  = librosa.feature.delta(mfcc)

chroma = librosa.feature.chroma_stft(S=log_S, sr=sr)

# melspectrogram
mel = librosa.feature.melspectrogram(y, sr=sr)

# spectral contrast
contrast = librosa.feature.spectral_contrast(S=log_S, sr=sr)

# We'll show each in its own subplot
plt.figure(figsize=(12, 2))

librosa.display.specshow(mfcc, sr=sr, x_axis='time')

marks = ['H', 'N']
for interval in humorous_tier: #humorous tier
    if interval.mark in marks:
        start = int(interval.minTime)
        end = int(interval.maxTime)
#         print(start,end)
        if(interval.mark == marks[0]):
            plt.axvspan(start, end, alpha=0.25, facecolor ='white', label='Humorous')
        else:
            plt.axvspan(start, end, alpha=0.25, facecolor ='black', label='Non-Humorous')
            
plt.ylabel('MFCC')
plt.colorbar()

plt.tight_layout()

# We'll show each in its own subplot
plt.figure(figsize=(12, 2))

librosa.display.specshow(delta_mfcc, sr=sr, x_axis='time')

marks = ['H', 'N']
for interval in humorous_tier: #humorous tier
    if interval.mark in marks:
        start = int(interval.minTime)
        end = int(interval.maxTime)
#         print(start,end)
        if(interval.mark == marks[0]):
            plt.axvspan(start, end, alpha=0.25, facecolor ='white', label='Humorous')
        else:
            plt.axvspan(start, end, alpha=0.25, facecolor ='black', label='Non-Humorous')
            
plt.ylabel('Delta MFCC')
plt.colorbar()

plt.tight_layout()

# We'll show each in its own subplot
plt.figure(figsize=(12, 2))

librosa.display.specshow(chroma, sr=sr, x_axis='time')

marks = ['H', 'N']
for interval in humorous_tier: #humorous tier
    if interval.mark in marks:
        start = int(interval.minTime)
        end = int(interval.maxTime)
#         print(start,end)
        if(interval.mark == marks[0]):
            plt.axvspan(start, end, alpha=0.25, facecolor ='white', label='Humorous')
        else:
            plt.axvspan(start, end, alpha=0.25, facecolor ='black', label='Non-Humorous')
            
plt.ylabel('Chroma')
plt.colorbar()

plt.tight_layout()
```
